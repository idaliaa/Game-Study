```python
#### Sentiment Per Comment ####
from textblob import TextBlob

# Load the data
df = pd.read_csv('cleaned_youtube_comments.csv')

# Function to get sentiment polarity (score between -1 and 1)
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity  # returns sentiment polarity (-1 to 1)

# Apply sentiment analysis to the 'comment' column
df['sentiment'] = df['comment'].apply(get_sentiment)

# Classify sentiment into Positive, Neutral, or Negative based on polarity
def classify_sentiment(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_label'] = df['sentiment'].apply(classify_sentiment)

# Print out a few rows to check the sentiment
print(df[['comment', 'sentiment', 'sentiment_label']].head())

# Plot the sentiment distribution
sentiment_counts = df['sentiment_label'].value_counts()

# Plotting the sentiment distribution
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'red', 'gray'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#### SENTIMENT ANALYSIS PER WORDS ####

# Function to get sentiment polarity for individual words
def get_word_sentiment(text):
    words = text.split()  # Split the comment into individual words
    word_sentiments = []  # List to store sentiment scores of each word
    for word in words:
        blob = TextBlob(word)  # Analyze the word's sentiment
        word_sentiments.append(blob.sentiment.polarity)  # Get polarity of each word
    return word_sentiments

# Apply word sentiment analysis to each comment and store the result in a new column
df['word_sentiment'] = df['comment'].apply(get_word_sentiment)

# Optional: Flatten the list of word sentiments for easier analysis
all_word_sentiments = [sentiment for sublist in df['word_sentiment'] for sentiment in sublist]

# Plot the sentiment distribution per word
plt.figure(figsize=(8, 6))
plt.hist(all_word_sentiments, bins=50, color='skyblue', edgecolor='black')
plt.title('Sentiment Distribution per Word')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()

# Print out the sentiment of the first few words for the first comment (as an example)
print("Word Sentiments for the first comment:")
print(df['word_sentiment'].iloc[0])
```
